```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df = pd.read_pickle("cleanDataDraft1.pkl")
```

```python
# df.info()
```

```python
df.dropna(inplace=True)
# df.info()
```

```python
X = df.drop("hiedinch", axis=1).values
Y = df[["hiedinch"]].values
```

```python
test = SelectKBest(score_func=chi2)
model = test.fit(X, Y)
np.set_printoptions(precision = 3)
score = model.scores_
features = model.transform(X)
columns = df.drop("hiedinch", axis=1).columns
```

```python
df_feature_score1 = pd.DataFrame({"Features": columns, "Score": score})
df_feature_score1.sort_values(by="Score", ascending=False)
```

Using PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=10)
model = pca.fit(X)
print("Explained Variances: %s" % model.explained_variance_ratio_)
```

Using Random Forest and Extra Trees

```python
from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier(n_estimators=10)
model.fit(X, Y)
score = model.feature_importances_
```

```python
df_feature_score2 = pd.DataFrame({"Features": columns, "Score": score})
df_feature_score2.sort_values(by="Score", ascending=False)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
model = LogisticRegression(solver="lbfgs")

rfe = RFE(model, 10)
fit = rfe.fit(X, Y)
# print(fit.n_features_)
# print(fit.support_)
rank = fit.ranking_
```

```python
df_feature_score3 = pd.DataFrame({"Features": columns, "Score": rank})
df_feature_score3.sort_values(by="Score", ascending=True)
```
